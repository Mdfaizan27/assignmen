```
[100, 2, 3]
```

### ğŸš« **Immutable Example: String**

```python
name = "Rahul"
name = "Amit"  # assigning a new string
print(name)
```

```
Amit
```

ğŸ‘‰ You didn't change the original string, you **created a new one**.  
So, **string is immutable**.

ğŸ‘‰ Output: `You are an adult`  
(Only runs if the condition is true.)


### âœ… 2. **if...else statement**

Used when you want to do one thing **if true**, and something else **if false**.

```python
age = 16

if age >= 18:
    print("You can vote")
else:
    print("You are too young to vote")
```

ğŸ‘‰ Output: `You are too young to vote`


### âœ… 3. **if...elif...else**

Used when you have **multiple conditions**.

```python
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 70:
    print("Grade B")
else:
    print("Grade C")
```

ğŸ‘‰ Output: `Grade B`

- Use `:` after `if`, `elif`, and `else`.
